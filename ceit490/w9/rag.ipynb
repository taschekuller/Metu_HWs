```python
!pip install langchain-huggingface
# Set your API key to a variable
import os
GROQ_API_KEY = ""

# Import the required packages
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from langchain_community.document_loaders import UnstructuredHTMLLoader
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.runnables import RunnablePassthrough
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
```

```python
# Load the HTML as a LangChain document loader
loader = UnstructuredHTMLLoader(file_path="data/mg-zs-warning-messages.html")
car_docs = loader.load()
```

```python
from langchain_core.prompts.chat import ChatMessage
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma



# Combine the car_docs into a single string for context
context = car_docs[0].page_content

# Split the context into smaller chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_text(context)


# Embed the chunks
embeddings = HuggingFaceEmbeddings(model_name="bert-base-uncased")
chunk_embeddings = embeddings.embed_documents(chunks)

# Create a Chroma vector store
chroma = Chroma.from_documents(chunks, chunk_embeddings)

# Define the user query
user_query = "What are the warning messages for the MG ZS?"

# Embed the user query
query_embedding = embeddings.embed_query(user_query)

# Retrieve the most relevant chunks
relevant_chunks = chroma.similarity_search(query_embedding, k=5)

# Combine the relevant chunks into a single string for context
relevant_context = "\n".join(relevant_chunks)

# Define the prompt
prompt = ChatPromptTemplate(
    messages=[
        ChatMessage(role="system", content="You are a helpful assistant."),
        ChatMessage(role="user", content=user_query)
    ],
    context=relevant_context
)

rag_chain = ({"context": relevant_context, "question": RunnablePassthrough()} | prompt | model)


# Run the prompt
response = prompt.invoke(prompt.messages)
print(response)
```
